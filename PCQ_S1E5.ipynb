```python
from qiskit import *
```

```python
circuit = QuantumCircuit(3,3)
```

```python
circuit.draw(initial_state=True, cregbundle=False)
```

```python
circuit.x(0)
circuit.barrier()
circuit.draw(initial_state=True, cregbundle=False)
```

```python
circuit.h(1)
circuit.cx(1,2)
```

```python
circuit.draw(initial_state=True, cregbundle=False)
```

```python
circuit.cx(0,1)
circuit.h(0)
circuit.draw(initial_state=True, cregbundle=False)
```

```python
circuit.barrier()
circuit.measure([0,1], [0,1])
circuit.draw(initial_state=True, cregbundle=False)
```

```python
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(initial_state=True, cregbundle=False)
```

```python
circuit.measure(2,2)
simulator =  Aer.get_backend('qasm_simulator')
print(simulator)
result = execute(circuit, backend = simulator, shot = 1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
